```python
from orbits.sim import SimRun
from astropy import units as u
from astropy.visualization import quantity_support
quantity_support()
from matplotlib import pyplot as plt
```

```python
run1 = SimRun(1*u.d,100*u.d)
```

```python
run1.add_object('test1',0.*u.au, 0.*u.au, 0.*u.au, 0*u.au/u.d, 0*u.au/u.d, 0*u.au/u.d, 1e30*u.kg, 0.01*u.au)
run1.add_object('test2',1.0*u.au, 0*u.au, 0*u.au, 0*u.au/u.d, -0.001*u.au/u.d, 0*u.au/u.d, 1e20*u.kg, 0.015*u.au)
run1.add_object('test3',2.0*u.au, 6*u.au, 0*u.au, 0.2*u.au/u.d, -0.3*u.au/u.d, 0*u.au/u.d, 1e22*u.kg, 0.015*u.au)
```

```python
#run1.objects
```

```python
run1.run()
```
