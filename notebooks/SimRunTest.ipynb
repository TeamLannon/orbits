```python
from orbits.sim import SimRun
from astropy import units as u
from astropy.visualization import quantity_support
quantity_support()
from matplotlib import pyplot as plt
```

```python
run1 = SimRun(1*u.d,10*u.d)
```

```python
run1.add_object('test1',0.5*u.au, 0.6*u.au, 0.7*u.au, 1.2*u.au/u.d, 1.3*u.au/u.d, 1.4*u.au/u.d, 100.*u.kg, 0.01*u.au)
run1.add_object('test2',3.5*u.au, 3.6*u.au, 3.7*u.au, 0.2*u.au/u.d, -0.3*u.au/u.d, 0.4*u.au/u.d, 150.*u.kg, 0.015*u.au)
```

```python
run1.objects
```

```python
run1.run()
```

```python
run1.objects
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(run1.objects[0]['pvt']['p'][:,0],run1.objects[0]['pvt']['p'][:,1],label=run1.objects[0]['name'])
ax.plot(run1.objects[1]['pvt']['p'][:,0],run1.objects[1]['pvt']['p'][:,1],label=run1.objects[1]['name'])
ax.legend()
```
